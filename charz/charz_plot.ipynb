```python
import skrf as rf
%matplotlib inline
from pylab import *
rf.stylely()
```

```python
minus_31p5dB = rf.Network('-31.5db.s2p')
minus_25p5dB = rf.Network('-25.5db.s2p')
minus_20p5dB = rf.Network('-20.5db.s2p')
minus_16p5dB = rf.Network('-16.5db.s2p')

minus_31p5dB.frequency = rf.Frequency(1,4400,501,"MHz")
minus_25p5dB.frequency = rf.Frequency(1,4400,501,"MHz")
minus_20p5dB.frequency = rf.Frequency(1,4400,501,"MHz")
minus_16p5dB.frequency = rf.Frequency(1,4400,501,"MHz")
```

```python
minus_16p5dB.plot_s_smith(m=0,n=0, label = "S11 @ Control word = -16.5dB")
minus_20p5dB.plot_s_smith(m=0,n=0, label = "S11 @ Control word = -20.5dB")
minus_25p5dB.plot_s_smith(m=0,n=0, label = "S11 @ Control word = -25.5dB")
minus_31p5dB.plot_s_smith(m=0,n=0, label = "S11 @ Control word = -31.5dB")
```

```python
minus_16p5dB.plot_s_db(m=1,n=0, label = "S21 Gain @ Control word = -16.5dB")
minus_20p5dB.plot_s_db(m=1,n=0, label = "S21 Gain @ Control word = -20.5dB")
minus_25p5dB.plot_s_db(m=1,n=0, label = "S21 Gain @ Control word = -25.5dB")
minus_31p5dB.plot_s_db(m=1,n=0, label = "S21 Gain @ Control word = -31.5dB")
```

```python
minus_16p5dB.plot_s_db(m=0,n=0, label = "S11 Return loss @ Control word = -16.5dB")
minus_20p5dB.plot_s_db(m=0,n=0, label = "S11 Return loss @ Control word = -20.5dB")
minus_25p5dB.plot_s_db(m=0,n=0, label = "S11 Return loss @ Control word = -25.5dB")
minus_31p5dB.plot_s_db(m=0,n=0, label = "S11 Return loss @ Control word = -31.5dB")
```
